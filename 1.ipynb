$$
\begin{cases}
{(y-y_1)\over(y_3-y_1)} = {(x-x_1)\over(x_3-x_1)} => y = {(x-x_1)(y_3-y_1)\over(x_3-x_1)}+y_1 \\
{(y-y_2)\over(y_4-y_2)} = {(x-x_2)\over(x_4-x_2)} => y = {(x-x_2)(y_4-y_2)\over(x_4-x_2)}+y_2 
\end{cases}
$$
${(x-x_1)(y_3-y_1)\over(x_3-x_1)}+y_1={(x-x_2)(y_4-y_2)\over(x_4-x_2)}+y_2$

$(x-x_1)(y_3-y_1)(x_4-x_2)+y_1(x_4-x_2)(x_3-x_1)=(x-x_2)(y_4-y_2)(x_3-x_1)+y_2(x_3-x_1)(x_4-x_2)$

$x={-x_2(y_4-y_2)(x_3-x_1)+y_2(x_3-x_1)(x_4-x_2)+x_1(y_3-y_1)(x_4-x_2)-y_1(x_4-x_2)(x_3-x_1)\over(y_3-y_1)(x_4-x_2)-(y_4-y_2)(x_3-x_1)}=>
y = {({-x_2(y_4-y_2)(x_3-x_1)+y_2(x_3-x_1)(x_4-x_2)+x_1(y_3-y_1)(x_4-x_2)-y_1(x_4-x_2)(x_3-x_1)\over(y_3-y_1)(x_4-x_2)-(y_4-y_2)(x_3-x_1)}-x_1)(y_3-y_1)\over(x_3-x_1)}+y_1$

```python
points = [[3, 5], [4, -2], [0, -4], [-2, 3]]
x1, y1 = points[0][0], points[0][1]
x2, y2 = points[1][0], points[1][1]
x3, y3 = points[2][0], points[2][1]
x4, y4 = points[3][0], points[3][1]
fig, ax = plt.subplots()
plt.grid()
try:
    if (x1 == x3) or (x2 == x4) or (y1 == y3) or (y2 == y4):
        raise ZeroDivisionError("Деление на ноль")
    x = ((-x2*(y4-y2)*(x3-x1)+y2*(x3-x1)*(x4-x2) +
         x1*(y3-y1)*(x4-x2)-y1*(x4-x2)*(x3-x1)) /
         ((y3-y1)*(x4-x2)-(y4-y2)*(x3-x1)))
    y = (x-x1)*(y3-y1)/(x3-x1)+y1
except ZeroDivisionError:
    print("Проверьте точки. Деление на 0!")
else:
    print([x, y])
    ax.plot(x, y, 'bo')
finally:
    ax.plot([x1, x2, x3, x4, x1], [y1, y2, y3, y4, y1])
    ax.plot([x1, x3], [y1, y3])
    ax.plot([x2, x4], [y2, y4])
```

```python
class vallis:
    def __init__(self, maxNumberIterations, n=-1):
        self.maxNumberIterations = maxNumberIterations
        self.n = n

    def __iter__(self):
        return self

    def __next__(self):
        self.n += 1
        if self.n >= self.maxNumberIterations:
            raise StopIteration
        return (2*self.n+2)**2/((2*self.n+1)*(2*self.n+3))
```

```python
res = 1
for x in vallis(1000):
    res *= x
print(res * 2)
```
